```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calculate the total number of schools

total_number_schools = len(school_data_complete["School ID"].unique())
total_number_schools
```

```python
# calculate the total number of students

total_number_students = len(school_data_complete["Student ID"].unique())
total_number_students
```

```python
#Calculate the total budget

total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculate the average maths score

average_maths_score = student_data["maths_score"].mean()
average_maths_score
```

```python
#Calculate the average reading score

average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

overall_average_score = (average_maths_score + average_reading_score)/2
overall_average_score
```

```python
#Calculate the percentage of students with a passing math score (50 or greater)

student_data["passing_maths"] = student_data["maths_score"] >= 50

percent_passing_maths = ((student_data["passing_maths"]).mean())*100
percent_passing_maths
```

```python
#Calculate the percentage of students with a passing reading score (50 or greater)

student_data["passing_reading"] = student_data["reading_score"] >= 50

percent_passing_reading = ((student_data["passing_reading"]).mean())*100
percent_passing_reading

```

```python
#Calculate the percentage of students who passed maths and reading (% Overall Passing)

overall_passing_rate = (percent_passing_maths + percent_passing_reading)/2
overall_passing_rate
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting
LGA_results = [{"Total Schools": total_number_schools, 
            "Total Students": total_number_students, 
            "Total Budget": total_budget, 
            "Avg Math Score":  round(average_maths_score,2), 
            "Avg Reading Score":  round(average_reading_score,2), 
            "Percent Passing Maths": round(percent_passing_maths,2),
            "Percent Passing Reading": round(percent_passing_reading,2),
            "Percent Overall Passing Rate": round(overall_passing_rate,2)}]
LGA_summary_table = pd.DataFrame(LGA_results)

#Formatting
LGA_summary_table["Percent Passing Maths"] = LGA_summary_table["Percent Passing Maths"].map("{:,.2f}%".format)
LGA_summary_table["Percent Passing Reading"] = LGA_summary_table["Percent Passing Reading"].map("{:,.2f}%".format)
LGA_summary_table["Percent Overall Passing Rate"] = LGA_summary_table["Percent Overall Passing Rate"].map("{:,.2f}%".format)
LGA_summary_table["Total Budget"] = LGA_summary_table["Total Budget"].map("${:,.2f}".format)
LGA_summary_table["Total Students"] = LGA_summary_table["Total Students"].map("{:,}".format)

#Display
LGA_summary_table
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Use groupby by school_name

school_name_group = school_data_complete.groupby(["school_name"]).mean()
school_name_group["Per Student Budget"] = school_name_group["budget"]/school_name_group["size"]
school_name_group["% Passing Maths"] = round(school_name_group["maths_score"]*100,2)
school_name_group["% Passing Reading"] = round(school_name_group["reading_score"]*100,2)
school_name_group["% Overall Passing"] = round(((school_name_group["maths_score"] + school_name_group["reading_score"])/2)*100,3)

#Merge with school_data to collect information about the type, size and budget
school_data_summary = pd.merge(school_name_group, school_data, how="left", on=["school_name", "school_name"])
del school_data_summary['size_y']
del school_data_summary['budget_y']
del school_data_summary['Student ID']
del school_data_summary['School ID_x']

school_summary_dataframe = pd.DataFrame({"School Name":  school_data_summary["school_name"],
                                "School Type": school_data_summary["type"],
                               "Total Students":school_data_summary["size_x"],
                               "Total School Budget": school_data_summary["budget_x"],
                               "Per Student Budget":school_data_summary["Per Student Budget"], 
                               "Average Maths Score":round(school_data_summary["maths_score"],2),
                               "Average Reading Score":round(school_data_summary["reading_score"],2), 
                               "% Passing Maths": school_data_summary["% Passing Maths"]/100,
                               "% Passing Reading": school_data_summary["% Passing Reading"]/100,
                               "% Overall Passing": school_data_summary["% Overall Passing"]/100}) 

#Formatting
school_summary_dataframe["Total Students"] = school_summary_dataframe["Total Students"].map("{:,.0f}".format)
school_summary_dataframe["Total School Budget"] = school_summary_dataframe["Total School Budget"].map("${:,.2f}".format)
school_summary_dataframe["Per Student Budget"] = school_summary_dataframe["Per Student Budget"].map("${:,.2f}".format)
school_summary_dataframe["% Passing Maths"] = school_summary_dataframe["% Passing Maths"].map("{:,.2f}".format)
school_summary_dataframe["% Passing Reading"] = school_summary_dataframe["% Passing Reading"].map("{:,.2f}".format)
school_summary_dataframe["% Overall Passing"] = school_summary_dataframe["% Overall Passing"].map("{:,.2f}".format)

#Display
school_summary_dataframe

```
